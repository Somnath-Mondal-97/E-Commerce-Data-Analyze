# Count the number of orders placed in 2017

```python
from SQLHelper import run_query
import pandas as pd
from IPython.display import display

query = 'SELECT COUNT(*) AS Total_Order_2017 FROM orders WHERE year(order_purchase_timestamp) = 2017'
df = run_query(query)
pd.set_option("display.max_rows",None)
display(df)
```

# Find the total sales per category

```python
from SQLHelper import run_query
import pandas as pd
from IPython.display import display

query = """SELECT products.[product category], COUNT(order_items.order_id) AS 'Total Product by Category'
FROM order_items JOIN products
ON order_items.product_id = products.product_id
GROUP BY products.[product category];"""
df = run_query(query)
pd.set_option("display.max_rows",None)
display(df)
```

# Calculate the percentage of orders that were paid in installments

```python
from SQLHelper import run_query
import pandas as pd
from IPython.display import display

query = """SELECT 
    CAST(
        (COUNT(CASE WHEN payment_installments > 1 THEN 1 END) * 100.0) / COUNT(*) 
        AS DECIMAL(18,2)
    ) AS Installment_Percentage
FROM payments;"""
df = run_query(query)
pd.set_option("display.max_rows",None)
display(df)
```

# Count the number of customers from each state.

```python
from SQLHelper import run_query
import pandas as pd
from IPython.display import display

query = """SELECT customer_state, COUNT(*)
FROM customers
GROUP BY customer_state;"""
df = run_query(query)
pd.set_option("display.max_rows",None)
display(df)
```
